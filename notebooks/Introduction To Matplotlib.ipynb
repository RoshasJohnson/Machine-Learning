## Matplotlib Workflow

```python
# 0.import matplotlib and get it ready for plotting in jupyter notebook
import matplotlib.pyplot as plt
```

```python
# 1.Prepare data
x = [1,2,4,5]
y = [11,22,33,44]

# 2.setup plot
fig, ax = plt.subplots(figsize = [10,10])

# 3.Plot data
ax.plot(x,y)

# 4.Customize plote 
ax.set(title = "Sample Plot",
      xlabel="x-axis",
      ylabel="y-axis")

# 5.save & show (you save the whole project figure)
fig.savefig("matplotlib/sample-plot.png")
```

 ## Making figures with NumPy arrays
 We want:
 * Line plot
 * Scatter plot
 * Bar plot
 * Histogram
 * Subplots

```python
import numpy as np
```

```python
# Create some data
x = np.linspace(0,10,100)
x[:10]
```

```python
# Plot the data create a line plot
fig,ax = plt.subplots()
ax.plot(x,x**2);
```

```python
fig,ax = plt.subplots()
ax.scatter(x,np.exp(x));
```

```python
 fig,ax = plt.subplots()
ax.scatter(x,np.sin(x));
```

```python
# Make a plot from dictionary 
nut_butter_prices = {"Almond Butter":10,
                    "Peanut Butter":8,
                    "Cashew Butter":12}
fig,ax = plt.subplots()
ax.bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax.set(title = "Roshas's Nut  Butter Store",ylabel="Price($)");
```

```python
# Make some data for histograms and plot it 
x = np.random.randn(1000)
fig,ax = plt.subplots()
ax.hist(x);
```

## Two optional for subplots

```python
# subplots option 1 
fig,((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows = 2,ncols = 2,figsize=(10,5))


# Plot to each different axis 
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# Subplots option 2 

fig, ax = plt.subplots(nrows = 2,ncols = 2,figsize=(10,5))
ax [1,0].plot(x,x/2)
```
